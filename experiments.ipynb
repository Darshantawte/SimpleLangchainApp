FAISS (Facebook AI Similiarity Search) is a library for efficient similarity search and clustering of dense vectors. It contains algorithms that search in sets of vectors of any size, up to ones that possibly do not fit in RAM. It also contains supporting code for evaluation and parameter tuning.

```python
from langchain_community.document_loaders import  TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import  OllamaEmbeddings
from langchain_text_splitters import  RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
db=FAISS.from_documents(docs,embeddings)
db
```

```python
#Query the FAISS vectordb
query="What is the reason speaker believe United States should enter the war?"
result = db.similarity_search(query=query)
```

```python
result[0].page_content
```

Retriever

Convert the vectorstoredb in a retriever class. Allows use it in other langchain methods, which largely work on retrievers

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
```

```python
#Similiarity Search with Score

docs_with_score = db.similarity_search_with_score(query=query)
docs_with_score
```

```python
#Saving vectorstoresb and load
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index",embeddings=embeddings,allow_dangerous_deserialization=True)
```

Chromadb

AI native opensource vector database

```python
from langchain_chroma import Chroma
from langchain_community.embeddings import  OllamaEmbeddings
from langchain_text_splitters import  RecursiveCharacterTextSplitter
from langchain_community.document_loaders import  TextLoader
```

```python
loader = TextLoader("speech.txt")
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
vectordb=Chroma.from_documents(documents=docs,embedding=embeddings)
```

```python
query="What is the reason speaker believe United States should enter the war?"
result = vectordb.similarity_search(query=query)
result
```
